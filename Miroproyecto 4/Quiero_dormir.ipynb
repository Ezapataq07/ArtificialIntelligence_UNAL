## Crear la lista de categorías desde el código

Suponiendo que cada clase está en un folder marcado con el nombre de la clase podemos aprovechar esto para no tener que crear la lista de categorías de forma manual

```python
import os
from tqdm import tqdm
```

```python
# ruta al directorio que contiene los folders
directorio = '/content/drive/Shareddrives/pruebas/'
```

```python
# leo nombres de folders y almaceno en lista
# también sirve con archivos
categorias = []
for folder in tqdm(os.listdir(directorio)):
  categorias.append(folder)
```
